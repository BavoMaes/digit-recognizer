```python
# Import modules
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
import math
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
import random
```

```python
# Read training data
raw_data = pd.read_csv("train.csv")
print(raw_data)

# Convert to numpy array
data = raw_data.to_numpy()
```

```python
# Split the data in test data and train data
train_data = data[0:21000,1:]
test_data = data[21000:, 1:]

train_labels = data[0:21000, 0]
test_labels = data[21000:, 0]

# Calculate the accuracy of a prediction
def calculate_accuracy(prediction):
    count = 0
    for i in range(0, 21000):
        count += 1 if prediction[i] == test_labels[i] else 0
    print("Accuracy =", round((count/21000)*100, 2), "%")
    
# Decision tree on first half and testing on second half
decision_tree = DecisionTreeClassifier(splitter='best')
decision_tree.fit(train_data, train_labels)

prediction = decision_tree.predict(test_data)
calculate_accuracy(prediction)
```

```python
# Calculate accuracy
def class_accs(predicted_labels, test_labels):
    indices0 = []
    indices1 = []
    for i, label in enumerate(test_labels):
        if label == 0:
            indices0.append(i)
        else:
            indices1.append(i)
    accuracy0 = metrics.accuracy_score([0] * len(indices0), predicted_labels[indices0])
    accuracy1 = metrics.accuracy_score([1] * len(indices1), predicted_labels[indices1])
    return accuracy0, accuracy1
```

```python
# Decision tree, 3 folds, criterion = "gini" (default), splitter = "best" (default)
folds = 3
bal_accs = []
decision_data = data[0:21000]
random.seed(1)
random.shuffle(decision_data)

for i in range(folds):
    start_fold = math.floor((i * (len(decision_data) / folds)))
    end_fold = math.floor(((i + 1) * (len(decision_data) / folds)))
    
    print(start_fold, "-", end_fold)
    
    training_fold = np.append(decision_data[:start_fold, 1:], decision_data[end_fold:, 1:], axis=0)
    training_labels = np.append(decision_data[:start_fold, 0], decision_data[end_fold:, 0], axis=0)
    
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(training_fold, training_labels)
    
    test_fold = decision_data[start_fold: end_fold, 1:]
    test_labels = decision_data[start_fold: end_fold, 0]
    
    prediction = decision_tree.predict(test_fold)
    
    acc0, acc1 = class_accs(prediction, test_labels)
    bal_acc = (acc0 + acc1) / 2.
    bal_accs.append(bal_acc)
print(bal_accs)
print('Balanced accuracy: {}'.format(np.mean(bal_accs)))
```
